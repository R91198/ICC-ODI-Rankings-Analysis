```python
import pandas as pd
import numpy as np
import time
import requests
import sys
from bs4 import BeautifulSoup
import re
```

```python
dates = pd.date_range(start= '2016-01-01',end = '2020-12-31', normalize= True).to_pydatetime().tolist()
dates = [d.strftime('%Y-%m-%d') for d in dates]
print(len(dates))
date = [i.split('-') for i in dates]
```

```python
batsmen=[]
batsmen_teams = []
batsmen_ratings = []

for i in date:
    url_odi_batsmen = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting?at={}-{}-{}'.format(i[0], i[1], i[2])
    req_data = requests.get(url_odi_batsmen)
    soup = BeautifulSoup(req_data.content, 'html.parser')

    
    #Batsmen Names
    top_rank = soup.findAll('div', {'class': 'rankings-block__banner--name-large'})
    for i in top_rank:
        name = i.text
        batsmen.append(name)
    players = soup.findAll('td', {'class':'table-body__cell rankings-table__name name'})
    for player in players:
        names = player.text.replace("\n", "")
        batsmen.append(names)
        
    #Batsmen Teams
    top_country = soup.find_all('div', {'class':'rankings-block__banner--nationality'})
    for country in top_country:
        top_name = country.text.replace('\n', '')
        batsmen_teams.append(top_name)
    countries = soup.findAll('span', {'class': 'table-body__logo-text'})
    for c in countries:
        country_names = c.text.replace("\n", "")
        batsmen_teams.append(country_names)
        
    #Batsmen Ratings
    top_rating = soup.find_all('div', {'class': 'rankings-block__banner--rating'})
    for r in top_rating:
        val = r.text
        batsmen_ratings.append(val)
    rest_rating = soup.find_all('td', {'class': 'table-body__cell rating'})
    for r in rest_rating:
        values = r.text
        batsmen_ratings.append(values)            
```

```python
print(len(batsmen))
print(batsmen[:10])
print()
print(batsmen_teams[:10])
print()
print(batsmen_ratings[:10])
```

```python
bowlers = []
bowler_teams = []
bowler_ratings = []

for i in date:
    url_bowler = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling?at={}-{}-{}'.format(i[0], i[1], i[2])
    req_data = requests.get(url_bowler)
    soup = BeautifulSoup(req_data.content, 'html.parser')

    #Bowlers Names
    top_rank = soup.findAll('div', {'class': 'rankings-block__banner--name-large'})
    for i in top_rank:
        name = i.text
        bowlers.append(name)
    players = soup.findAll('td', {'class':'table-body__cell rankings-table__name name'})
    for player in players:
        names = player.text.replace("\n", "")
        bowlers.append(names)
        
    #Bowlers Teams    
    top_country = soup.find_all('div', {'class':'rankings-block__banner--nationality'})
    for c in top_country:
        top_name = c.text.replace('\n', '')
        bowler_teams.append(top_name)
    countries = soup.findAll('span', {'class': 'table-body__logo-text'})
    for c in countries:
        country = c.text.replace("\n", "")
        bowler_teams.append(country)
        
    #Bowlers Ratings
    top_rating = soup.find_all('div', {'class': 'rankings-block__banner--rating'})
    for r in top_rating:
        val = r.text
        bowler_ratings.append(val)
    rest_rating = soup.find_all('td', {'class': 'table-body__cell rating'})
    for r in rest_rating:
        values = r.text
        bowler_ratings.append(values) 
```

```python
print(len(bowlers))
print(bowlers[:10])
print()
print(bowler_teams[:10])
print()
print(bowler_ratings[:10])
```

```python
all_rounders = []
all_round_teams = []
all_round_ratings = []

for i in date:
    url_all_rounder = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/all-rounder?at={}-{}-{}'.format(i[0], i[1], i[2])
    req_data = requests.get(url_all_rounder)
    soup = BeautifulSoup(req_data.content, 'html.parser')

    #All-rounders Names
    top_rank = soup.findAll('div', {'class': 'rankings-block__banner--name-large'})
    for i in top_rank:
        name = i.text
        all_rounders.append(name)
    players = soup.findAll('td', {'class':'table-body__cell rankings-table__name name'})
    for player in players:
        names = player.text.replace("\n", "")
        all_rounders.append(names)
        
    #All-rounder Teams
    top_country = soup.find_all('div', {'class':'rankings-block__banner--nationality'})
    for c in top_country:
        top_name = c.text.replace('\n', '')
        all_round_teams.append(top_name)
    countries = soup.findAll('span', {'class': 'table-body__logo-text'})
    for c in countries:
        country = c.text.replace("\n", "")
        all_round_teams.append(country)
    
    #All-rounder Ratings
    top_rating = soup.find_all('div', {'class': 'rankings-block__banner--rating'})
    for r in top_rating:
        val = r.text
        all_round_ratings.append(val)
    rest_rating = soup.find_all('td', {'class': 'table-body__cell rating'})
    for r in rest_rating:
        values = r.text
        all_round_ratings.append(values)       
```

```python
print(len(all_rounders))
print(all_rounders[:10])
print()
print(all_round_teams[:10])
print()
print(all_round_ratings[:10])
```

```python
#Dataframe for batsmen
data_batsmen = {'Player Names': batsmen, 'Team': batsmen_teams, 'Rating': batsmen_ratings, 'Type': 'Batsmen'}
df_batsmen = pd.DataFrame(data_batsmen)

#Dataframe for bowlers
data_bowler = {'Player Names': bowlers, 'Team': bowler_teams, 'Rating': bowler_ratings, 'Type': 'Bowler'}
df_bowler = pd.DataFrame(data_bowler)

#Dataframe for all-rounders
data_all_round = {'Player Names': all_rounders, 'Team': all_round_teams, 'Rating': all_round_ratings, 'Type': 'All-rounder'}
df_all_rounder = pd.DataFrame(data_all_round)
```
